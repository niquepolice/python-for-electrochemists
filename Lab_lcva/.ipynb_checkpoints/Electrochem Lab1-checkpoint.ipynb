Похоже на обычную csv таблицу, только с ';' в качестве разделителя. csv удобно загружать с помощью pandas -  библиотеки для работы с таблицами

```python
import pandas as pd
# функция, которая строит циклы по данным из файла
def plot_vac(filename):
    df = pd.read_csv('data/' + filename, sep=';') # импортируем данные в dataframe (pandas-таблица) из csv-файла
    df.columns = ['Scan', 'Index', 'Time', 'E', 'I'] # меняем названия полей на более удобные

    for i in df.Scan.unique():
        scan = df[df.Scan == i]
        plt.title(filename)
        plt.plot(scan.E, scan.I, label=i)
    plt.legend()
```

```python
plt.figure(figsize=(9,5)) # размер графика
plot_vac('e1 Pt 1mm BG CV')
```

```python
plt.figure(figsize=(9,5))
plot_vac('e2 Pt 1mm BG CV')
```

```python
plt.figure(figsize=(9,5))
plot_vac('e3')
```

```python
plt.figure(figsize=(18,10))
for i in range(4):
    fn = ['1', '2', '5_1', '5_2'][i]
    plt.subplot('22' + str(i+1))
    plot_vac(fn)
```

```python
plt.figure(figsize=(18,10))
for i in range(3):
    fn = ['0.1', '0.2', '0.5'][i]
    plt.subplot('22' + str(i+1))
    plot_vac(fn)
```

Построим зависимость тока пика (для первого цикла) от квадратного корня из скорости развёртки, сделаем линейную регрессию

```python
from scipy.stats import linregress
v = np.array([0.1, 0.2, 0.5, 1, 2, 5])
I_max = []
for fn in [0.1, 0.2, 0.5, 1, 2, "5_1"]:
    df = pd.read_csv('data/' + str(fn), sep=';')
    df.columns = ['Scan', 'Index', 'Time', 'E', 'I']
    I_max.append(df[df.Scan == 1].I.values.max())
x = v**0.5
plt.scatter(x, I_max)
k, c, r_value, p_value, err = linregress(x, I_max)
plt.plot(x, x * k + c)
plt.ylim(0, 4e-5)
plt.title(r'$I_{max} \left(\sqrt{\dfrac{dE}{dT}}\right)$')
print('Коэффициент детерминации: ', r_value**2)
```
