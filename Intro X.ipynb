```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
# без %matplotlib inline картинка может не выводиться
n = 100
x = np.linspace(1, 10, n) # n равноудалённых друг от друга точек на отрезке [1, 10]
plt.plot(x, np.sin(x), label='sin') # добавляем ломаную на график 
plt.plot(x, np.sqrt(x), label='sqrt')
plt.legend()
plt.title("two curves")
plt.show() # опционально
```

```python
from scipy import misc
img = misc.face(gray=True) # загрузим енота в виде двумерного числового массива
plt.figure(figsize=(17,6))
plt.subplot(121)
plt.imshow(img, cmap='gray')
# правильный способ посчитать градиент конечными разностями в питоне 
# (или произвести любую другую операцию с числовыми массивами)
# - через numpy. Циклами получится раз в 100 медленнее (numpy - сишный модуль и работает быстро)
def grad(img):
    gradx = (np.roll(img, -1, axis=0) - img)
    grady = (np.roll(img, -1, axis=1) - img)
    
    return np.stack((gradx, grady))
%time g = grad(img)
plt.subplot(122)
plt.imshow(g[0]**2 + g[1]**2, cmap='gray')
```

```python
from IPython.display import Audio
rate, f, temp = 44100, 440, 100
def note(shift, sound, pause):
    sound, pause = sound * 60 / temp , pause * 60 / temp
    ticks = lambda t : int(rate*t)
    sin = np.sin(np.linspace(0, np.pi*f*2**(shift/12)*sound, ticks(sound)))
    zero = np.zeros(ticks(pause))
    return np.hstack((sin, zero))

notes = [note(sh,so, p) for sh, so, p in 
             [
              (-2,0.8,0.2),(-2,0.8,0.2),(-2,0.8,0.2),
              (-6,0.5,0.25),(1,0.25,0),(-2,0.8,0.2),
              (-6,0.5,0.25),(1,0.25,0),(-2,1,1),

              (5,0.8,0.2),(5,0.8,0.2),(5,0.8,0.2),
              (6,0.5,0.25),(1,0.25,0),(-3,0.8,0.2),
              (-6,0.5,0.25),(1,0.25,0),(-2,1,1)
             ]
        ]
Audio(np.hstack(notes), rate=rate)
```
